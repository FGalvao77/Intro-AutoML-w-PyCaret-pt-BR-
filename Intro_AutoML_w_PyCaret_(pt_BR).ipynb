````python
# importando a biblioteca
import researchpy as rp 
````

````python
# criando uma funo para auxiliar na visualizao das avaliaes estatsticas e das imagens grficas
def categorical_analysis(data, index, col, figsize=(7, 5)):
    abs_tbl, test = rp.crosstab(data[index], data[col], 
                                test='chi-square', margins=False)
    row_tbl = rp.crosstab(data[index], data[col], prop='row')
    row_tbl = row_tbl.iloc[:-1, :-1]
    
    abs_tbl = abs_tbl.sort_values(('left', 1))
    row_tbl = row_tbl.sort_values(('left', 1))
    
    print(test)
    print('\n')
    
    _, ax = plt.subplots(1, 2, figsize=figsize, sharey=True)
    
    sns.heatmap(abs_tbl, annot=True, fmt='.5g', cbar=False, cmap='Blues', 
                annot_kws={'fontsize': 20}, ax=ax[0])
    sns.heatmap(row_tbl, annot=True, fmt='.5g', cbar=False, cmap='Greens', 
                annot_kws={'fontsize': 20}, ax=ax[1])
    plt.show();
````

**Teste qui-quadrado de Pearson**

O `teste qui-quadrado (2) de Pearson` (ou `teste chi-quadrado de Pearson`)  um teste estatstico aplicado a dados categricos para avaliar quo provvel  que qualquer diferena observada acontea ao acaso.  adequado para amostras no pareadas/emparelhadas. Dentre os diversos testes qui-quadrado existentes (por exemplo, a _correo de continuidade de Yates_, _teste da razo da mxima verossimilhana_, _teste de Portmanteau em sries temporais_, etc)  o mais utilizado.

(...)

Testa-se a [hiptese nula](https://pt.wikipedia.org/wiki/Hip%C3%B3tese_nula) afirmando-se que a [distribuio de frequncias](https://pt.wikipedia.org/wiki/Distribui%C3%A7%C3%A3o_de_frequ%C3%AAncias) de um certo evento observado em uma [amostra](https://pt.wikipedia.org/wiki/Amostra_(estat%C3%ADstica))  consistente com uma distribuio terica particular. Os eventos considerados devem ser mutuamente excludentes e devem ter probabilidade total 1. Um caso comum para este teste  quando os eventos cobrem um valor de sada de uma varivel categrica. Um simples exemplo  a hiptese de que um [dado](https://pt.wikipedia.org/wiki/Dado) de 6 lados  "honesto" (isto , todos os sei possveis valores - 1, 2, 3, 4, 5 e 6 - so equiprovveis).

- Fonte: https://pt.wikipedia.org/wiki/Teste_qui-quadrado_de_Pearson

De forma mais simplista.

O Teste qui-quadrado  um teste no-paramtrico que serve para comparar duas propores quando os dados assumirem qualquer distribuio. Tambm testa o grau de associao entre as variveis.

Para interpretar um teste Chi Quadrado de Pearson  preciso observar o p-valor resultante e tambm a tabela para verificarmos onde est essa diferena.

A interpretao segue de maneira bem simples. Considerando que p-valores abaixo de 0.05 so significativo, se o seu resultado  significativo  preciso observar a tabela e verificar onde est essa diferena. Caso contrrio, dizemos que no h evidncias o suficiente para mostrar a no homoneidade da tabela.

**Referncias teis**
- [Wikipedia](https://en.wikipedia.org/wiki/Chi-squared_test#Pearson's_chi-squared_test) sobre o teste Chi Quadrado de Pearson (em ingls)
- Livro Estatstica Bsica: Bussab, Wilton O., and Pedro A. Morettin. Estatstica bsica. So Paulo: Atual, 1986.

**Valor-p**

Na [estatstica](https://pt.wikipedia.org/wiki/Estat%C3%ADstica) clssica, o `valor-p` (tambm chamado de `nvel descritivo` ou `probabilidade de significncia`),  a [probabilidade](https://pt.wikipedia.org/wiki/Probabilidade) de se obter uma estatstica de teste igual ou mais extrema que aquela observada em uma amostra, sob a hiptese nula. Por exemplo, em testes de hiptese, pode-se rejeitar a [hiptese nula](https://pt.wikipedia.org/wiki/Hip%C3%B3tese_nula) a 5% caso o `valor-p` seja _menor que 5%_. Assim, uma outra interpretao para o `valor-p`,  que este  o menor nvel de significncia com que se rejeitaria a hiptese nula. Em termos gerais, um `valor-p` pequeno significa que a probabilidade de obter um valor da estatstica de teste como o observado  muito improvvel, levando assim  rejeio da hiptese nula.

Em um teste clssico de hipteses, so definidas duas hipteses, a _nula (Ho)_ e a _alternativa (Ha)_. Em muitas aplicaes da estatstica, convenciona-se definir a hiptese alternativa como a hiptese formulada pelo pesquisador, enquanto a hiptese nula  o seu complemento. A princpio, a hiptese nula  considerada a verdadeira. Ao confrontarmos a hiptese nula com os achados de uma amostra aleatria tomada de uma populao de interesse, verifica-se a sua plausibilidade em termos probabilsticos, o que nos leva a rejeitarmos ou no Ho. Se no rejeitamos Ho, tomamos como sendo verdadeira; caso contrrio, tomamos Ha como verdadeira.

- Fonte: https://pt.wikipedia.org/wiki/Valor-p

Se o valor-p for menor que 0.05, devemos rejeitar a hiptese nula de que no h diferena entre as mdias e concluir que existe uma diferena significativa. Se o valor-p for maior que 0.05, no  possvel concluir que existe uma diferena significativa.

    P-value            |  Interpretao
    -------------------|-----------------------------------------
    P < 0,01	       |  evidncia muito forte contra H0
    0,01 < = P < 0,05  |  evidncia moderada contra H0
    0,05 < = P < 0,10  |  evidncia sugestiva contra H0
    0,10 < = P	     |  pouca ou nenhuma evidncia real contra H0


**V-quadrado de Cramer**

_Os valores altos do V de Cramer indicam uma relao mais forte entre as variveis, e os valores menores para o V indicam uma relao fraca._
 - valor prximo de `0` indica que **no existe uma associao**. 
 - valor prximo de `1` indica que **no h uma associao muito forte entre as variveis**.

- Fonte: https://en.wikipedia.org/wiki/Cram%C3%A9r's_V

**Matriz de confuso**

Em anlise preditiva, a `matriz de confuso` (s vezes tambm chamada de _'matriz de erro'_ ou _'tabela de confuso'_)  uma tabela com duas linhas e duas colunas que relata o nmero de `falsos positivos`, `falsos negativos`, `verdadeiros positivos` e `verdadeiros negativos`.

No campo do **Aprendizado de Mquina** uma _matriz de confuso_  uma tabela que permite a visualizao do desempenho de um algoritmo de classificao
- Fonte: https://pt.wikipedia.org/wiki/Matriz_de_confus%C3%A3o

````python
# importando a funo
from sklearn.metrics import classification_report, plot_confusion_matrix
````

````python
# iremos criar uma funo para otimizar a aplicao do modelo
#  e a exibio do relatrio e da matriz de confuso
def classification_results(classifier, x_test, y_test, cmap='Blues'):
    print('Resultados do modelo:', classifier.__class__.__name__)
    
    y_pred = classifier.predict(X_test)
    
    print(classification_report(y_test, y_pred))
    
    plot_confusion_matrix(classifier, X_test, y_test, cmap=cmap);
````

````python
# aplicando a funo criada e passando como argumentos o modelo 
# e as bases de testes - "X_test" e "y_test"
classification_results(lr, X_test, y_test)
````

> Podemos visualizar que, para essa aplicao o modelo apresentou uma baixa performance.
